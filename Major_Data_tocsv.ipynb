```python
workforce_demographic2018['OCC_GROUP'] = workforce_demographic2018['OCC_GROUP'].astype(str)
detailed_groups_df = workforce_demographic2018.loc[workforce_demographic2018.OCC_GROUP.apply(lambda cat: 'detailed' in cat)]
major_groups_df = workforce_demographic2018.loc[workforce_demographic2018.OCC_GROUP.apply(lambda cat: 'major' in cat)]
#detailed_groups_raw_data = detailed_groups_df.drop(['AREA','STATE','ST','OCC_GROUP','OCC_TITLE','ANNUAL','HOURLY'], axis=1)
```

```python
major_groups_clean = major_groups_df.drop(['AREA','STATE','ST','OCC_GROUP','ANNUAL','HOURLY','H_PCT90','A_PCT90','OCC_CODE'], axis=1)
```

```python
major_groups_clean.head()
```

```python
major_groups_numeric = major_groups_clean.replace(',','',regex=True)
```

```python
major_groups_numeric['OCC_TITLE'] = major_groups_numeric['OCC_TITLE'].astype(str)
```

```python
major_groups_numeric = major_groups_numeric.iloc[:,1:].astype(float)
```

```python
major_groups_numeric.astype(float);
```

```python
major_groups_numeric.to_csv('major_groups_unnormalized.csv', index=False)
```

```python
from sklearn import preprocessing

x = major_groups_numeric.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

major_data_normalized = pa.DataFrame(x_scaled)
```

```python
occ_titles = []
occ_titles = major_groups_clean.iloc[:,0:1]
```

```python
col_one_list = occ_titles['OCC_TITLE'].tolist()
```

```python
indexed_occ_title = pa.DataFrame(col_one_list) 
```

```python
major_data_normalized.insert(0, "OCC TITLE", indexed_occ_title)
```
